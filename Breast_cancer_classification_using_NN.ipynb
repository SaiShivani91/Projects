Building Neural Network

An NN consists of input and output layers. Between these input and output layers, there are several hidden layers. We are going to use Tensorflow and keras.

Tensorflow is a deep learning library developed by google. The most widely used deep learning libraries are tensorflow and pytorch.

Keras sets tensorflow or pytorch as backend to implement neural networks in a easy way.

```python
import tensorflow as tf
tf.random.set_seed(3) # in neural network several weights and parameters are created. Random seed is set because everytime we run code the same values are created and the accuracy does not chsnge . It is same as random state in train test split.
from tensorflow import keras
```

```python
# setting up the layers the neural network

model = keras.Sequential([keras.layers.Flatten(input_shape = (30,)),
keras.layers.Dense(20, activation = 'relu'),
keras.layers.Dense(2, activation = 'sigmoid')])
```

30 = Number of neurons = Number of features in input layer
20 = Number of neurons in hidden layer
2 = Number of neurons in output layer = Number of classes

Firing of neurons means the output will be given  by a single neuron. The other neurons are blacked

```python
# Each neuron in hidden layer will be having the activation function relu. If we want to add another hidden layer the code is keras.layers.Dense(30, activation = "sigmoid")
```

Dense means all the neurons are connected to each other and all the hidden layer neurons get the information from all the input neurons.

Flatten - X train will be in the form of matrix. But in NN, the data must be flatten to a 1D array so that each feature will go into one neuron. The number of neurons in input layer = the number of features in the dataset. This is specific for numerical data.

Sparase categorical cross entropy is used when we have categorical labels like dog = 0, cat = 1 and horse = 2.
Categorical entropy cross entropy is used when we have one hot encoding like dog = [1 0 0], Cat = [0 1 0], Horse = [0 0 1]

```python
# Compiling the neural network

model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Training the neural network

history = model.fit(X_train, Y_train, validation_split = 0.1, epochs = 10) # validation data = 10%, epoch = how many times the model has to go through the data.
```

Low accuracy is due to not standardizing the data.

```python
# Standardizing the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test) # should not fit X test data
```

```python
print(X_train_std)
```

```python
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 10)
```

Loss is difference between actual and predicted values. Loss is inversely proportional to Accuracy

```python
# Visualize Accuracy and Loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['training data', 'validation data'] , loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')

plt.legend(['training data', 'validation data'] , loc = 'upper right')
```

```python
# Accuracy of the model on test data

loss, accuracy = model.evaluate(X_test_std, Y_test)
```
